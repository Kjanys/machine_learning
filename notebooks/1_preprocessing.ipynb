## Предобработка датасета Zoo

```python
import pandas as pd

zoo_data = pd.read_csv('../data/zoo.csv')
zoo_data.head()
```

## Пропуски в данных

Проверка dataset Zoo на пропуски в данных

```python
zoo_data.isnull().sum()
```

Пропусков в данных нет, перейдем к перекодировке

## Признак animal-name

Так как все строки данного признака имеют уникальное значение и не влияют на построение модели имеет смысл удалить данный признак

```python
zoo_data = zoo_data.drop('animal-name', 1)
zoo_data.head()
```

## Сохранение в csv

```python
zoo_data.to_csv('../data/zoo_preprocessed.csv', index=False)
```

## Предобработка датасета Forest Fires

```python
forestfires_data = pd.read_csv('../data/forestfires.csv')
forestfires_data.head()
```

## Категориальные признаки

К категориальными признакам относятся month, day. Преобразуем их используя one-hot encoding

```python
forestfires_data = pd.get_dummies(forestfires_data,prefix=['month'], columns=['month'])
forestfires_data = pd.get_dummies(forestfires_data,prefix=['day'], columns=['day'])

forestfires_data.head()
```

## Сохранение в csv

```python
forestfires_data.to_csv('../data/forestfires_preprocessed.csv', index=False)
```
