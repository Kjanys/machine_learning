# Визуализация данных

## Датасет Zoo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
zoo_data = pd.read_csv('../data/zoo_preprocessed.csv')
zoo_data.head()
zoo_numeric_df = pd.concat([zoo_data['tail'], zoo_data['legs'], 
                       zoo_data['venomous'],zoo_data['predator']],
                       axis=1, keys=['tail', 'legs','venomous','predator'])
zoo_numeric_df.hist(bins=20, figsize=(20,20))
plt.show()
```

Сделанная выше визуализация показывает распределение наличия хвостов, количества ног, ядовитых и хищных животных относительно общего количества зверей в выборке.

Из графиков можно сделать следующие выводы. Более половины животных в выборке являются хищниками. Ядовитыми, являются менее 10% всех животных в выборке. 2/3 животных не имеют хвоста. Большинство животных выборки являются четвероногими.

## Датасет Forest Fires

```python
forest_data = pd.read_csv('../data/forestfires_preprocessed.csv')
forest_numeric_df = pd.concat([forest_data['area'], forest_data['rain'], 
                       forest_data['wind'],forest_data['temp']], axis=1, 
                       keys=['area', 'rain','wind','temp'])
forest_numeric_df.hist(bins=20, figsize=(20,20))
plt.show()
```

Сделанная выше визуализация показывает распределение площади пожаров, количтво осадков, скорость ветра и температуры территории относительно общего количества зон в выборке.

Из графиков можно сделать следующие выводы. Большая часть пожаров имеет малую площадь. Большая часть пожаров возникает на территориях с малым колличством осадков. Большая часть пожаров происходят при скорости ветра до 6 км/ч. Пик пожаров приходится на среднюю температуру в 20 градусов цельсия.
